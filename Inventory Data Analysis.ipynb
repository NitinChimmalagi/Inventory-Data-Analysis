```python
# Set figure size
plt.figure(figsize=(15, 5))

# Get top 10 products/brands with highest prices
vc = PurchasePrice.groupby('Description')['Price'].max().sort_values(ascending=False)[:10]

# Create barplot
g = sns.barplot(x=vc.index, y=vc.values, palette="viridis")

# Annotate each bar with its value
for i, value in enumerate(vc.values):
    g.text(x=i, y=value - 2, s=round(value, 2), color='black', ha='center', fontsize=10)

# Add plot title and labels
plt.title('Top 10 Brands with Highest Price Tags')
plt.xlabel('Brands')
plt.ylabel('Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(12, 5))

# Top 10 vendors with highest total amount
vc = Invoice.groupby('VendorName')['Total Amount'].max().sort_values(ascending=False)[:10]

# Barplot (no need to include `data=Invoice` here)
g = sns.barplot(x=vc.index, y=vc.values, color='orange')

# Annotate bars
for i, value in enumerate(vc.values):
    g.text(x=i, y=value - 2, s=round(value, 2), color='black', ha='center', fontsize=8)

# Titles and labels
plt.title('Top 10 Vendors with Highest Amount Paid')
plt.xlabel('Vendor Names')
plt.ylabel('Amount')
plt.xticks(rotation=35)

# Clean layout
plt.tight_layout()
plt.show()
```

```python
Final_Sales['Total Amount'] = Final_Sales['SalesDollars'] + Final_Sales['ExciseTax']
Final_Sales.head(2)
```

```python
# Set the figure size
plt.figure(figsize=(12, 5))

# Get top 10 vendors by total sales amount
vc = Final_Sales.groupby(['Description'])['Total Amount'].sum().sort_values(ascending=False)[:10]

# Barplot without passing the entire DataFrame
g = sns.barplot(x=vc.index, y=vc.values, palette="Set2")

# Annotate each bar with the sales value
for i, value in enumerate(vc.values):
    g.text(x=i, y=value - 2, s=round(value, 2), ha='center', color='black', fontsize=10)

# Add titles and labels
plt.title('Top 10 Vendors with Highest Sale Amount')
plt.xlabel('Brands')
plt.ylabel('Total Sale Amount')
plt.xticks(rotation=25)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 5))

# Get top 10 brands by maximum Excise Tax
vc = Final_Sales.groupby('Description')['ExciseTax'].max().sort_values(ascending=False)[:10]

# Barplot without 'data=' since we're passing x and y directly
g = sns.barplot(x=vc.index, y=vc.values, palette="Set2")

# Annotate each bar with tax value
for i, value in enumerate(vc.values):
    g.text(x=i, y=value - 2, s=round(value, 2), ha='center', color='black', fontsize=10)

# Set chart title and labels
plt.title('Top 10 Brands with Highest Excise Tax')
plt.xlabel('Brands')
plt.ylabel('Excise Tax')
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Top 10 vendors by maximum sales quantity
vc = Final_Sales.groupby('VendorName')['SalesQuantity'].max().sort_values(ascending=False)[:10]

# Barplot without passing the full DataFrame
g = sns.barplot(x=vc.index, y=vc.values, hue=vc.index, palette="Set1", legend=False)


# Annotate the bars
for i, value in enumerate(vc.values):
    g.text(x=i, y=value - 2, s=round(value, 2), ha='center', color='black', fontsize=10)

# Set title and labels
plt.title('Top 10 Vendors with Highest Sale Quantity')
plt.xlabel('Vendors')
plt.ylabel('Sale Quantity')
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```
